```python
#number of female athelets in olympics

female_participants = atheltes_df[(atheltes_df.Sex == 'F') & (atheltes_df.Season == 'Summer')][['Sex' , 'Year']] 
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
WomenOlympics = atheltes_df[(atheltes_df.Sex == 'F') & (atheltes_df.Season== 'Summer')]
WomenOlympics
```

```python
sns.set(style="darkgrid")
plt.figure(figsize = (20,10))
sns.countplot(x= 'Year', data = WomenOlympics , palette = 'Spectral')
plt.title('Women particpation')
```

```python
part = WomenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:,'F'].plot()
plt.title('plot of female Athelets over time')
```

```python
Gold_Meadls = atheltes_df[(atheltes_df.Medal == 'Gold')]
Gold_Meadls
```

```python
Gold_Meadls = Gold_Meadls[np.isfinite(Gold_Meadls['Age'])]
```

```python
Gold_Meadls['ID'][Gold_Meadls['Age'] > 60].count()
```

```python
sports_event = Gold_Meadls['Sport'][Gold_Meadls['Age']>60]
sports_event
                                                            
```

```python
#plot for sporting_event

plt.figure(figsize = (10,5))
plt.tight_layout()
sns.countplot( x = sports_event , data = sports_event)
plt.title('Gold Medals for Athelets Over 60 Years')
```

```python
# Gold Medals from Each Country 

Gold_Meadls.Region.value_counts().reset_index(name = 'Medal').head(5)
```

```python
total_gold_medals = Gold_Meadls.Region.value_counts().reset_index(name = 'Medal').head(6)
g = sns.catplot(x = 'index' , y = 'Medal' , data = total_gold_medals , height = 5 , kind = "bar" , palette = "rocket")
g.despine(left = True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
```
