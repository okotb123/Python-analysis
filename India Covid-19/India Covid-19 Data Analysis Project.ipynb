```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational" , "ConfirmedForeignNational"] , inplace = True , axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# Active Cases

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed" , "Cured" , "Deaths"] , 
                           index = "State/UnionTerritory" , aggfunc = max)
statewise
```

```python
statewise['recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Death Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = 'cubehelix')
```

```python
# top 10 active states 

top_10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases' , 'Date']].sort_values(by = ['Active_Cases'],ascending =False).reset_index()
```

```python
top_10_active_cases
```

```python
top_10_active_cases 
fig = plt.figure(figsize=(16,9))
plt.title( 'Top 10 states with most active  cases in india' , size = 20)
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active_Cases" , x = "State/UnionTerritory" , linewidth = 2 , edgecolor = 'red')
plt.xlabel("states")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# top states with high death
top_10_death = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths' ,'Date']].sort_values(by = ['Deaths'] ,ascending = False).reset_index()
```

```python
top_10_death
fig = plt.figure(figsize = (16,5))
plt.title('Top 10 states with most Death in india' , size = 20)
ax=sns.barplot(data = top_10_death.iloc[:12], y = 'Deaths' , x = 'State/UnionTerritory' ,linewidth= 2 , edgecolor = 'black')
plt.xlabel("states")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
fig = plt.figure(figsize = (12,6))
ax = sns.lineplot( data = covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra" , "Karnataka" , "Kerala" , "Tamil Nadu" ,"Uttar Pradesh"])] ,x = 'Date' , y = 'Active_Cases' , hue = 'State/UnionTerritory')
ax.set_title("Top 5 Affected States in India" , size = 16)
                  
```

```python
#remove sows where state = India

vaccine = vaccine_df[vaccine_df.State != 'India']
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated" : "Total"}, inplace = True)
vaccine.head()
```

```python
#Most Vaccintaed State

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total' ,ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (10,5))
plt.title('Top 5 states Vaccintaed in india' , size = 20)
ax=sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total , x = max_vac.index ,linewidth= 2 , edgecolor = 'black')
plt.xlabel("states")
plt.ylabel("Vaccintation")
plt.show()
```
